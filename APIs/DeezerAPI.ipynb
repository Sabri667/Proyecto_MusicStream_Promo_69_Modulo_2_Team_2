### Consulta de tracks (2 resultados incompletos Taylor S y TINI)

```python
import requests
import time

def consulta_tracks(limit_por_artista, delay):
    lista_artistas = [
        "Taylor Swift", "Drake", "Morgan Wallen", "Kendrick Lamar", "The Weeknd",
        "SZA", "Zach Bryan", "Tyler the Creator", "Billie Eilish", "Ariana Grande",
        "Bad Bunny", "Karol G", "Myke Towers", "Quevedo", "Aitana",
        "Rosalía", "Morad", "Melendi", "Enrique Iglesias", "Juanes",
        "Duki", "María Becerra", "Tini Stoessel", "Nicki Nicole", "Cazzu",
        "Esposito", "Paulo Londra", "Shakira", "Bizarrap", "Emilia",
    ]
    
    #headers para evitar bloqueos
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    
    tracks_search = {'id': [], 'title': [], 'rank': [], 'artist': []}
    
    for artista in lista_artistas:
        url= f"https://api.deezer.com/search/track?q=artist:\"{artista}\"&index=1&limit={limit_por_artista}"
        #url_1= f"https://api.deezer.com/search/track?q=artist:\"{artista}\"&limit={limit_por_artista}"
        print(f"Consultando tracks para: {artista}")
        
        try:
            respuesta = requests.get(url, headers=headers)
            if respuesta.status_code != 200:
                print(f"Error {respuesta.status_code} para {artista}")
                continue
            
            datos = respuesta.json()
            tracks = datos.get('data', [])
            
            if not tracks:              #en primer consulta arrojó 2 canciones para Taylor S.
                print(f"No se encontraron tracks para {artista}")
                continue
            
            for track in tracks:            #ver de agregar variables
                tracks_search['id'].append(track['id'])
                tracks_search['title'].append(track['title'])
                tracks_search['rank'].append(track.get('rank', 0))
                tracks_search['artist'].append(artista)  # guardamos el nombre buscado
            
            print(f"  → {len(tracks)} tracks obtenidos de {artista}")
            
        except Exception as e:
            print(f"Excepción con {artista}: {e}")
        
        # tiempo de espera para no sobrecargar la API
        time.sleep(delay)
    
    return tracks_search

# llamado y conversión a tabla
resultado = consulta_tracks(limit_por_artista=100, delay=1)
print(f"Total de tracks recolectados: {len(resultado['id'])}")

import pandas as pd
df = pd.DataFrame(resultado)
print(df.head())

```

## Consulta de canciones por Id artista (RUTA:TOP)

```python
import requests
import time

def consulta_cancion(lista_artistas):
    info_canciones = []
       
    for artista in lista_artistas:
        id_artista = artista["id_artista"]
        nombre_artista = artista["nombre_artista"]

        consulta_canciones = requests.get(f"https://api.deezer.com/artist/{id_artista}/top?limit=50")

        if consulta_canciones.status_code != 200:
            print(f"No se pudo consultar las canciones del artista: {nombre_artista}")
            continue
        datos_canciones = consulta_canciones.json()

        for cancion in datos_canciones["data"]:
            info_canciones.append({"id_artista": id_artista,"nombre_artista": nombre_artista,"id_cancion": cancion["id"],
                            "titulo_cancion": cancion["title"],"id_album":cancion ["album"]["id"]})#,"titulo_album": cancion["album"]["title"]})#,
                            #"contributors": cancion.get("contributors")})

            
        time.sleep(0.3)
    return info_canciones
```

## Consulta de album por Id Canción (RUTA: ALBUM)

```python
def consulta_album(info_cancion):
    id_canciones = []
    info_album= []
    for cancion in info_cancion:
        id_cancion = cancion["id_cancion"]
        titulo_cancion = cancion["titulo_cancion"]
        id_album = cancion["id_album"]

        consulta_album = requests.get(f"https://api.deezer.com/album/{id_album}")

        if consulta_album.status_code != 200:
            print(f"No se pudo consultar la canción {id_cancion}: {titulo_cancion}.")
            continue

        datos_album = consulta_album.json()

        info_album.append({'id_cancion':{id_cancion}, 'titulo_cancion':{titulo_cancion}, 'id_album':{id_album},
                          'id_genre':datos_album['id_genre'],
                          'genre': datos_album['genres']['data']['name'], 
                          'año_de_lanzamiento':datos_album['release_date']})
    return info_album
    
    
```
