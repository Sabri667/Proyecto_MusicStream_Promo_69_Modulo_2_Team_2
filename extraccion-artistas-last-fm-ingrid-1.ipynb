```python
info_artistas = []

for nombre_artista in artistas:

    params = {
        "method": "artist.getInfo",
        "artist": nombre_artista,
        "api_key": api_key,
        "format": "json",
        "lang": "es",
        "autocorrect": 1                # para corregir nombrede artista q esté mal escrito y lo traiga bien aunque no lo tengo en la lista de artistas, pero por si acaso.
    }

    respuesta = requests.get(url, params=params)
    datos = respuesta.json()

    if "artist" not in datos:
        print(f"Artista no encontrado: {nombre_artista}")
        continue

    artista = datos["artist"]

    # similares
    similares = []
    for similar in artista["similar"]["artist"]:
        similares.append(similar["name"])

    # tags (añadimos esto ahora) -para ver estilo(pop), genero mujer, etc. - es una lista de etiquetas que los usuarios han asignado al artista en Last.fm.
    tags = []
    for tag in artista["tags"]["tag"]:
        tags.append(tag["name"])

    # biografía limpia
    bio_limpia = artista["bio"]["summary"].split("<a href=")[0].strip() # lo limpia para quedarnos solo con el resumen sin el enlace a la bio completa

    # guardo todo en un diccionario y lo añado a la lista
    info_artistas.append({
        "artista_buscado": nombre_artista,
        "nombre": artista["name"],
        "listeners": artista["stats"]["listeners"],
        "playcount": artista["stats"]["playcount"],
        "biografia": bio_limpia,
        "similares": similares,
        "tags": tags
    })

    time.sleep(0.3)
```

```python
df_lastfm = pd.DataFrame(info_artistas)
df_lastfm
```

```python
df_lastfm["listeners"] = df_lastfm["listeners"].astype(int)
df_lastfm["playcount"] = df_lastfm["playcount"].astype(int)
```

```python
df_lastfm["ratio_popularidad"] = df_lastfm["playcount"] / df_lastfm["listeners"]
```
