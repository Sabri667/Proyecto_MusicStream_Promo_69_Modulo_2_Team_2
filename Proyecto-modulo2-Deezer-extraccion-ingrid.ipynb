#### Consulta de artistas en Deezer API

```python
lista_artistas = [
 
    # 🇺🇸 EEUU (10)
    "Taylor Swift", "Drake", "Morgan Wallen", "Kendrick Lamar", "The Weeknd",
    "SZA", "Zach Bryan", "Tyler the Creator", "Billie Eilish", "Ariana Grande",
 
    # 🇪🇸 España (10)
    "Bad Bunny", "Karol G", "Myke Towers", "Quevedo", "Aitana",
    "Rosalía", "Morad", "Melendi", "Enrique Iglesias", "Juanes",
 
    # 🇦🇷 Argentina (10)
    "Duki", "María Becerra", "Tini Stoessel", "Nicki Nicole", "Cazzu",
    "Lali", "Paulo Londra", "Shakira", "Bizarrap", "Emilia",
]

artista_search = []

for artista in lista_artistas:
    consulta=requests.get(f"https://api.deezer.com/search/artist?q={artista}&index=0&limit=1")
    consulta.status_code

    if consulta.status_code != 200:
        print(f"No se pudo realizar la consulta del artista: {artista}.\n Error : {consulta.status_code}") 

    datos=consulta.json()
    artista_search.append(datos['data'][0])

print(artista_search)
```

LISTA INFORMACION DE ARTISTAS

```python
artistas_info = []
for artista in artista_search:
    artistas_info.append({"id_artista": artista["id"], "nombre_artista":artista["name"]})
    for artista in artistas_info:     
        if artista["nombre_artista"] == "Morgan Wallen":         
            artista["id_artista"] = 7188840
        if artista["nombre_artista"] == "Shakira":         
            artista["id_artista"] = 160

artistas_info
    
```

LISTA CANCIONES

```python
import time

canciones = []
for artista in artistas_info:
    id_artista = artista["id_artista"]
    nombre_artista = artista["nombre_artista"]

    consulta_canciones = requests.get(f"https://api.deezer.com/artist/{id_artista}/top?limit=50")

    if consulta_canciones.status_code != 200:
        print(f"No se pudo consultar las canciones del artista: {nombre_artista}")
        continue
    datos_canciones = consulta_canciones.json()

    for cancion in datos_canciones["data"]:
        canciones.append({"id_artista": id_artista,
            "nombre_artista": nombre_artista,
            "titulo_cancion": cancion["title"],
            "titulo_album": cancion["album"]["title"],
            "tipo_cancion": cancion["type"]})
        
    time.sleep(1)
    
```
